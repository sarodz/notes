Inverse Iteration

```octave
N = size(A,1);
v = randn(N,1);
v = v/norm(v);

mu = 0

for e=1:100,
    w = (A-mu*eye(N))\v;
    v = w/norm(w);    
    l = v'*A*v
    %mu = l;
end
```
